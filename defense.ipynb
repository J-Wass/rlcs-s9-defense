```python
import os, sys, traceback, time

import carball
from carball.json_parser.game import Game
from carball.analysis.analysis_manager import AnalysisManager
from carball import decompile_replays

# List of analysis_manager proto messages for each .replay file.
analysis_list = []

# Dict of .replay file paths to their errors (if found).
errors = {}

# List of paths of each .replay file.
replay_path_list = []

# Directory to walk for replays.
target_directory = "./replays"

def decompile(filepath):
    """Returns an analysis manager object for a .replay file path.
    Args:
        filepath (string): The .replay file path to decompile.
        analysis_list (multiprocessing.queue): The queue to put the decompiled object into.
    """
    json = decompile_replays.decompile_replay(filepath,
                                    output_path=filepath.replace(".replay",".json"),
                                    overwrite=True)
    game = Game()
    game.initialize(loaded_json=json)
    analysis_manager = AnalysisManager(game)
    analysis_manager.create_analysis(calculate_intensive_events=True)
    return analysis_manager

# Navigate local replays directory for all .replay files.
start = time.time()
for root, dirs, files in os.walk(target_directory):
    for file in files:
        if file.endswith(".replay"):
            filepath = os.path.join(root, file)
            try:
                analysis_list.append(decompile(filepath))
            except Exception as e:
                print("Could decompile {0} :(".format(filepath))
                errors[filepath] = e

# Elapsed time, in minutes.
elapsed_time = round((time.time() - start)/60)
print("Decompiled {0} replays in {1} minutes.".format(len(analysis_list), elapsed_time))
    
```

```python
def get_boost_steals_allowed(proto, defending_team, game_teams):
    # Generate all player ids.
    defending_player_ids = set(defending_team.team_player_ids)
    all_player_ids = set()
    for t in game_teams:
        for pl_id in t.player_ids:
            all_player_ids.add(pl_id.id)

    # Generate opponent hits from all player ids.
    opponent_player_ids = all_player_ids-defending_player_ids
    opponent_players = list(filter(lambda x: x.id.id in opponent_player_ids, proto.players))
    opponent_stolen_boosts = sum(map(lambda x: x.stats.boost.num_stolen_boosts, opponent_players))
    return opponent_stolen_boosts

def get_shots_allowed(proto, defending_team):
    defending_player_ids = set(defending_team.team_player_ids)
    hits = proto.game_stats.hits
    shots_allowed = 0
    for hit in hits:
        if hit.shot and hit.player_id.id not in defending_player_ids:
            shots_allowed += 1
    return shots_allowed

def get_saves(proto, defending_team):
    defending_player_ids = set(defending_team.team_player_ids)
    hits = proto.game_stats.hits
    saves = 0
    for hit in hits:
        if hit.save and hit.player_id.id in defending_player_ids:
            saves += 1
    return saves

def get_clears(proto, defending_team):
    defending_player_ids = set(defending_team.team_player_ids)
    hits = proto.game_stats.hits
    clears = 0
    for hit in hits:
        if hit.clear and hit.player_id.id in defending_player_ids:
            clears += 1
    return clears

def get_clear_distance(proto, defending_team):
    """Return the total distance covered in clears."""
    defending_player_ids = set(defending_team.team_player_ids)
    hits = proto.game_stats.hits
    clear_disance = 0
    for hit in hits:
        if hit.clear and hit.player_id.id in defending_player_ids:
            clear_disance += hit.distance
    return clear_disance

def get_total_defensive_shot_pressure(proto, defending_team, game_teams):
    """Determine the average hit pressure that teams apply when 
    defending. Also populates defending_team.hit_pressure_map.
    
    Args:
       proto (carball.api.Game): The proto that represents the game.
       defending_team (carball.api.Team): The defending team to calculate pressure.
       game_teams (List[local class Team]): The list of all teams involved in this game.
    """
    # Generate all player ids.
    defending_player_ids = set(defending_team.team_player_ids)
    all_player_ids = set()
    for t in game_teams:
        for pl_id in t.player_ids:
            all_player_ids.add(pl_id.id)

    # Generate opponent hits from all player ids.
    opponent_player_ids = all_player_ids-defending_player_ids
    opponent_shots = list(filter(lambda x: x.shot and (x.player_id.id in opponent_player_ids), proto.game_stats.hits))
    
    # Get a sample defender to determine if defense is orange.
    defending_ids = all_player_ids-opponent_player_ids
    sample_defending_player_id = next(iter(defending_ids))
    sample_defending_player = next(filter(lambda x: x.id.id == sample_defending_player_id,proto.players))
    
    # Filter opponent hits by hits on defending half.
    defense_is_orange = sample_defending_player.is_orange
    if (defense_is_orange):
        opponent_offensive_hits = list(filter(lambda x: x.ball_data.pos_y > 0, opponent_shots))
    else:
        opponent_offensive_hits = list(filter(lambda x: x.ball_data.pos_y < 0, opponent_shots))
        
    # Calculate hit pressure graph.
    for shot in opponent_shots:
        x = shot.ball_data.pos_x
        y = shot.ball_data.pos_y
        if (defense_is_orange):
            y *= -1
        shot_pressure_map = defending_team.shot_pressure_map
        shot_pressure_map[(x,y)] = shot.pressure
        
    # Return the average pressure of only the defending hits.
    opponent_shot_pressure = sum((map(lambda x: x.pressure, opponent_offensive_hits)))
    return opponent_shot_pressure
```

```python
import copy
import pickle

def analysis_list_to_tuples(analysis_list):
    analysis_tuples = []
    for analysis in analysis_list:
        proto_string = analysis.protobuf_game.SerializeToString()
        dataframe = analysis.data_frame
        analysis_tuple = (dataframe, proto_string)
        analysis_tuples.append(analysis_tuple)
    return analysis_tuples

def save_analysis_list(analysis_list):
    analysis_tuples = analysis_list_to_tuples(analysis_list)
    with open("analysis_list", "wb+") as f:
        pickle.dump(analysis_tuples, f) 

def load_analysis_tuples():
    with open("analysis_list", "rb") as f:
        return pickle.load(f)

# Save the analysis list so we don't have to do that again.
#save_analysis_list(analysis_list)
# Load it into tuples.
analysis_list = load_analysis_tuples()
print(len(analysis_list))
```

```python
import carball

def normalized_team_name(team_name):
    """Normalize the team name to always be the same."""
    if team_name == "BARCA ESPORTS":
        return "BARÇA ESPORTS"
    if team_name == "TEAM ENDPOINT":
        return "ENDPOINT"
    if team_name == "GHOST":
        return "GHOST GAMING"
    return team_name

eu = [
    "TEAM SINGULARITY",
    "TEAM RECIPROCITY",
    "RENAULT VITALITY",
    "ENDPOINT",
    "VELOCE ESPORTS",
    "BARÇA ESPORTS",
    "TSM",
    "DIGNITAS",
    "AS MONACO",
    "MOUSESPORTS",
]

# Dict of all teams in analysis_list with team_name as key.
teams = {}
class Team:
    def __init__(self, team_player_ids):
        self.team_player_ids = team_player_ids
        self.total_defensive_pressure = 0
        self.boost_steals_allowed = 0
        self.shots_allowed = 0
        self.games = 0
        self.clears = 0
        self.clear_distance = 0
        self.saves = 0
        # key is (x,y) coordinate and value is pressure
        self.shot_pressure_map = {}
        self.color = "Blue"
        
    def update_player_ids(self, new_ids):
        for player_id in new_ids:
            if player_id not in self.team_player_ids:
                self.team_player_ids.append(player_id)

# Populate the teams dict from analysis_list
for analysis in analysis_list:
    proto = carball.generated.api.game_pb2.Game()
    proto.ParseFromString(analysis[1])
    game_teams = proto.teams
    for team in game_teams: 
        # add team if it doesn't exist
        team_name = normalized_team_name(team.name)
        if team_name not in teams:
            team_player_ids = list(map(lambda x: x.id, team.player_ids))
            teams[team_name] = Team(team_player_ids)
            if team_name in eu:
                teams[team_name].color = "Blue"
            else:
                teams[team_name].color = "Red"
        else:
            teams[team_name].update_player_ids(list(map(lambda x: x.id, team.player_ids)))

        current_team = teams[team_name]
        current_team.games += 1

        current_team.total_defensive_pressure += get_total_defensive_shot_pressure(proto, current_team, game_teams)
        current_team.shots_allowed += get_shots_allowed(proto, current_team)
        current_team.boost_steals_allowed += get_boost_steals_allowed(proto, current_team, game_teams)
        current_team.clears += get_clears(proto, current_team)
        current_team.clear_distance += get_clear_distance(proto, current_team)
        current_team.saves += get_saves(proto, current_team)

import pickle
pickle.dump(teams, open("teams-all", "wb+"))
```
